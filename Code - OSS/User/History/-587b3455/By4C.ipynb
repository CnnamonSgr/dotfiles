```python
var("t") 
Xs = t^3+t^2+50


show(
plot(Xs, t, 0, 12, ymax=3000, ymin=0, axes_labels=['$t[h]$', '$x[km]$'], figsize = 10, legend_label="Samolot", color="red") +
point((2, 62), size = 25, color = 'orange', legend_label='2 Sekunda') +
point((4, 130), size = 25, color = 'orange', legend_label='4 Sekunda') +
point((8, 626), size = 25, color = 'orange', legend_label='8 Sekunda') 

)

pochodna=diff(Xs,t)
print(solve(pochodna,t))
print(f"Położenie w 2 sekundzie: {Xs(t=2)}")
print(f"Położenie w 4 sekundzie: {Xs(t=4)}")
print(f"Położenie w 8 sekundzie: {Xs(t=8)}")


Vs=diff(Xs,t)
print(f"\nPrędkość Samolotu = {Vs}")

show(
plot(Vs, t, 0, 12, ymax=3000, ymin=0, axes_labels=['$t[h]$', '$x[km/h]$'], figsize = 10, legend_label="Samolot", color="red") +
point((2, 16), size = 25, color = 'orange', legend_label='2 Sekunda') +
point((4, 56), size = 25, color = 'orange', legend_label='4 Sekunda') +
point((8, 208), size = 25, color = 'orange', legend_label='8 Sekunda') 
)

pochodna=diff(Vs,t)
print(solve(pochodna,t))
print(f"Prędkość w 2 sekundzie: Vs(t=2)")
print(f"Prędkość w 4 sekundzie: Vs(t=4)")
print(f"Prędkość w 8 sekundzie: Vs(t=8)")

As=diff(Vs,t)
print(f"Przyspieszenie Samolotu = {As}")


show(
plot(As, t, 0, 12, ymax=3000, ymin=0, axes_labels=['$t[h]$', '$x[km/h]$'], figsize = 10, legend_label="Samolot", color="red") +
point((2, 14), size = 25, color = 'orange', legend_label='2 Sekunda') +
point((4, 26), size = 25, color = 'orange', legend_label='4 Sekunda') +
point((8, 50), size = 25, color = 'orange', legend_label='8 Sekunda') 
)

pochodna=diff(As,t)
print(solve(pochodna,t))
print(f"Prędkość w 2 sekundzie: As(t=2)")
print(f"Prędkość w 4 sekundzie: As(t=4)")
print(f"Prędkość w 8 sekundzie: As(T=8)")
```
